```python
import pandas as pd


df =pd.read_csv("data/users.csv")
df
```

```python
filas = df.shape[0]
res = []
user=""
interes =""
users_intereses =df[["user_id", "interests"]]
lista_intereses =   [(1, "travel"), 
    (2,"finance"), 
    (3,"gaming"),
    (4, "lifestyle"),
    (5,"news"),
    (6, "food"),
    (7, "photography"),
    (8,"art"),
    (9,"sports"),
    (10, "health"),
    (11, "fashion"),
    (12, "technology"),
    (13, 'fitness')]
interes = users_intereses.at[0, "user_id"]
interes
for i in range(filas):
    user = users_intereses.at[i, "user_id"]
    interes = users_intereses.at[i, "interests"]
    valores = interes.split(', ')
   
    for j in range(len(valores)):
        interes_id =[tup for tup in lista_intereses if valores[j] in tup]
        res.append((user, interes_id[0][0]))

lista_sin_duplicados_ordenada = []
for elemento in res:
    if elemento not in lista_sin_duplicados_ordenada:
        lista_sin_duplicados_ordenada.append(elemento)
print(lista_sin_duplicados_ordenada)

```
